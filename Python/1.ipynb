## Python Variables

A variable is a name used to store data in memory. Python variables do not need explicit declaration. A variable is created when a value is assigned to it.

Rules for variable names:

- Must start with a letter or underscore
- Cannot start with a number
- Can contain letters, numbers, and underscores

# Unit II: Python Program Flow Control

**Syllabus topics:** Conditional blocks using `if`, `else`, `else if`; simple `for` loops; `for` loops using ranges, strings, lists and dictionaries; `while` loops; loop manipulation using `pass`, `continue`, `break` and `else`; programming using conditional and loop blocks.

## Conditional Blocks: if, else and elif

Conditional statements are used when a program needs to make decisions.

Python uses:

- `if` for the first condition
- `elif` for additional conditions. This is Python's version of else-if.
- `else` when no previous condition is true

````python
marks = 82

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
````

## Simple for Loops in Python

A `for` loop is used to iterate over a sequence such as a range, string, list, tuple, set, or dictionary.

Basic syntax:

```python
for variable in sequence:
    statement
```

````python
for number in range(1, 6):
    print(number)
````

````python
# for loop with string
for char in "PYTHON":
    print(char)

# for loop with list
subjects = ["Math", "Python", "DBMS"]
for subject in subjects:
    print(subject)

# for loop with dictionary
student = {"name": "Aman", "age": 20, "branch": "CSE"}
for key, value in student.items():
    print(key, "=", value)
````

## while Loops in Python

A `while` loop runs as long as a condition is true. It is useful when the number of repetitions is not fixed in advance.

Always update the loop variable inside the loop. Otherwise, the loop may become infinite.

````python
count = 1

while count <= 5:
    print(count)
    count += 1
````

## Loop Manipulation: pass, continue, break and else

- `pass`: does nothing. It is used as a placeholder.
- `continue`: skips the current iteration and moves to the next iteration.
- `break`: stops the loop immediately.
- loop `else`: runs when the loop finishes normally without `break`.

````python
for number in range(1, 8):
    if number == 2:
        pass
    if number == 3:
        continue
    if number == 6:
        break
    print(number)
else:
    print("Loop completed without break")
````

````python
# Program: print even and odd numbers from 1 to 10
for number in range(1, 11):
    if number % 2 == 0:
        print(number, "is even")
    else:
        print(number, "is odd")
````

## String Data Type and String Operations

A string is a sequence of characters. Strings are immutable, meaning once created, their content cannot be changed directly.

Common operations:

- indexing
- slicing
- concatenation
- repetition

````python
text = "Python Programming"

print(text[0])
print(text[-1])
print(text[0:6])
print(text + " Language")
print("Hi" * 3)
print("Python" in text)
````

## String Manipulation Methods

String methods are built-in functions for working with strings.

Useful methods:

- `upper()`
- `lower()`
- `title()`
- `strip()`
- `replace()`
- `split()`
- `join()`
- `find()`

````python
message = "  python is powerful  "

print(message.upper())
print(message.lower())
print(message.title())
print(message.strip())
print(message.replace("powerful", "easy"))
print("apple,banana,mango".split(","))
print("-".join(["2026", "04", "29"]))
print(message.find("python"))
print(message.count("p"))
````

## Defining List and List Slicing

A list stores multiple values in a single variable. Lists are ordered, mutable, and allow duplicate values.

List slicing is used to access a part of a list.

Syntax:

```python
list_name[start:stop:step]
```

````python
numbers = [10, 20, 30, 40, 50, 60]

print(numbers[0])
print(numbers[1:4])
print(numbers[:3])
print(numbers[3:])
print(numbers[::2])
print(numbers[::-1])
````

## List Manipulation

Because lists are mutable, we can add, remove, update, and sort items.

Common methods:

- `append()`
- `insert()`
- `remove()`
- `pop()`
- `sort()`
- `reverse()`
- `count()`
- `index()`

````python
fruits = ["apple", "banana", "mango"]

fruits.append("orange")
fruits.insert(1, "grapes")
print(fruits)

fruits.remove("banana")
print(fruits)

last_item = fruits.pop()
print(last_item)
print(fruits)

marks = [50, 90, 70, 85]
marks.sort()
print(marks)
marks.reverse()
print(marks)
````

## Tuple Data Type

A tuple is an ordered collection like a list, but it is immutable. Once a tuple is created, its elements cannot be changed.

Tuples are useful when data should remain fixed.

````python
student = (101, "Ravi", "CSE")

print(student)
print(student[0])
print(student[1])

roll, name, branch = student
print(roll, name, branch)
````

## Dictionary Manipulation

A dictionary stores data in key-value pairs. Keys are unique and are used to access values.

Common methods:

- `keys()`
- `values()`
- `items()`
- `get()`
- `update()`
- `pop()`

````python
student = {
    "name": "Aisha",
    "roll": 102,
    "branch": "IT",
    "marks": 91
}

print(student["name"])
print(student.get("marks"))
print(student.keys())
print(student.values())
print(student.items())

student["marks"] = 95
student.update({"city": "Lucknow"})
print(student)

student.pop("city")
print(student)
````

## Introduction to File Operations

File handling allows a program to store data permanently. Python provides the `open()` function to open a file.

Common file modes:

- `r`: read mode
- `w`: write mode, overwrites existing content
- `a`: append mode
- `x`: create mode
- `b`: binary mode
- `t`: text mode

Using `with open(...)` is recommended because it closes the file automatically.

````python
# Writing to a file using write()
with open("student.txt", "w") as file:
    file.write("Name: Rahul\n")
    file.write("Branch: CSE\n")
    file.write("Marks: 85\n")

print("File written successfully")
````

````python
# Reading complete file using read()
with open("student.txt", "r") as file:
    data = file.read()

print(data)
````

## Understanding read(), readline() and readlines()

- `read()` reads the entire file as one string.
- `readline()` reads one line at a time.
- `readlines()` reads all lines and returns a list of strings.

````python
with open("student.txt", "r") as file:
    first_line = file.readline()
    second_line = file.readline()

print("First line:", first_line)
print("Second line:", second_line)

with open("student.txt", "r") as file:
    lines = file.readlines()

print(lines)
````

## Understanding write() and writelines()

- `write()` writes a single string to a file.
- `writelines()` writes a list of strings to a file.

`writelines()` does not automatically add newline characters, so include `\n` where needed.

````python
lines = [
    "Python file operations\n",
    "write() writes one string\n",
    "writelines() writes multiple strings\n"
]

with open("notes.txt", "w") as file:
    file.writelines(lines)

with open("notes.txt", "r") as file:
    print(file.read())
````

## Manipulating File Pointer using seek()

A file pointer indicates the current position in a file. When data is read, the pointer moves forward.

`seek(position)` moves the file pointer to a specific position.

````python
with open("notes.txt", "r") as file:
    print(file.read(6))
    print("Current pointer position:", file.tell())
    file.seek(0)
    print("After seek:")
    print(file.read(6))
````

````python
# Program: count number of lines and words in a file
with open("notes.txt", "r") as file:
    lines = file.readlines()

line_count = len(lines)
word_count = 0

for line in lines:
    word_count += len(line.split())

print("Lines:", line_count)
print("Words:", word_count)
````

## NumPy Introduction

## GUI Programming with Tkinter

GUI means Graphical User Interface. Tkinter is Python's standard library for creating GUI applications.

Tkinter programs usually follow these steps:

1. Import Tkinter
2. Create the main window
3. Add widgets
4. Define events/functions
5. Run the main event loop using `mainloop()`

## Tk Widgets

Widgets are GUI components.

Common Tkinter widgets:

- `Label`: displays text
- `Button`: clickable button
- `Entry`: single-line text input
- `Text`: multi-line text area
- `Frame`: container for grouping widgets
- `Checkbutton`: checkbox
- `Radiobutton`: select one option
- `Listbox`: list of items

````python
# Tkinter example: basic window
# Run this in a local Python environment. It may not display inside some notebook setups.

import tkinter as tk

window = tk.Tk()
window.title("First Tkinter App")
window.geometry("300x150")

label = tk.Label(window, text="Hello Tkinter")
label.pack(pady=20)

button = tk.Button(window, text="Close", command=window.destroy)
button.pack()

window.mainloop()
````
